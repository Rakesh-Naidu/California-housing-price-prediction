## Data Wrangling
### Outliers Treatment

```python
sns.boxplot(x='MedInc',data=df,showmeans=True,meanline=True,showfliers=True)
```

```python
df['MedInc'] = np.sqrt(df['MedInc'])
q = df["MedInc"].quantile(0.98)
df = df[df["MedInc"]<q]
q = df["MedInc"].quantile(0.02)
df = df[df["MedInc"]>q]
```

```python
df.shape
```

```python
sns.boxplot(x='AveRooms',data=df,showmeans=True,meanline=True,showfliers=True)
```

```python
df['AveRooms'] = np.sqrt(df['AveRooms'])
q = df["AveRooms"].quantile(0.98)
df = df[df["AveRooms"]<q]
q = df["AveRooms"].quantile(0.02)
df = df[df["AveRooms"]>q]
```

```python
df.shape
```

```python
sns.boxplot(x='AveBedrms',data=df,showmeans=True,meanline=True,showfliers=True)
```

```python
df['AveBedrms'] = np.sqrt(df['AveBedrms'])
q = df["AveBedrms"].quantile(0.95)
df = df[df["AveBedrms"]<q]
q = df["AveBedrms"].quantile(0.05)
df = df[df["AveBedrms"]>q]
```

```python
df.shape
```

```python
sns.boxplot(x='Population',data=df,showmeans=True,meanline=True,showfliers=True)
```

```python
df['Population'] = np.sqrt(df['Population'])
q = df["Population"].quantile(0.98)
df = df[df["Population"]<q]
q = df["Population"].quantile(0.02)
df = df[df["Population"]>q]
```

```python
df.shape
```

```python
sns.boxplot(x='AveOccup',data=df,showmeans=True,meanline=True,showfliers=True)
```
